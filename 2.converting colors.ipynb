```python
import cv2
from matplotlib import pyplot as plt

# Read image
img = cv2.imread("wallp.png")

# Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img_rgb.shape)
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray.shape)
# Create figure with 2 plots
plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

# Grayscale Image
plt.subplot(1,2,2)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")

plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("wallp.png")

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)
lab = cv2.cvtColor(rgb, cv2.COLOR_RGB2LAB)
xyz = cv2.cvtColor(rgb, cv2.COLOR_RGB2XYZ)

plt.figure(figsize=(15,8))

plt.subplot(2,3,1)
plt.imshow(rgb)
plt.title("RGB")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(gray, cmap="gray")
plt.title("Gray")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(hsv)
plt.title("HSV")
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(lab)
plt.title("LAB")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(xyz)
plt.title("XYZ")
plt.axis("off")

plt.show()
```
