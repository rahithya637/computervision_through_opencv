```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def show(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```

```python
import cv2
import numpy as np

img = np.zeros((400,400,3), dtype=np.uint8)

cv2.line(
    img,            # image
    (50,50),        # start point
    (350,350),      # end point
    (0,255,0),      # color (B,G,R)
    5               # thickness
)

show(img)
```

```python
img = np.zeros((400,400,3), dtype=np.uint8)

cv2.rectangle(
    img,
    (100,100),      # top-left
    (300,250),      # bottom-right
    (255,0,0),      # blue
    3
)

show(img)
```

```python
img = np.zeros((400,400,3), dtype=np.uint8)

cv2.rectangle(img,(100,100),(300,250),(255,0,0),-1)

show(img)
```

```python
img = np.zeros((400,400,3), dtype=np.uint8)

cv2.circle(
    img,
    (200,200),      # center
    80,             # radius
    (0,0,255),      # red
    4
)

show(img)
```

```python
img = np.zeros((400,400,3), dtype=np.uint8)

cv2.circle(img,(200,200),80,(0,0,255),-1)

show(img)
```

```python
img = np.zeros((400,400,3), dtype=np.uint8)

cv2.ellipse(
    img,
    (200,200),      # center
    (120,60),       # axes
    0,              # angle
    0,              # start angle
    360,            # end angle
    (255,255,0),
    3
)

show(img)
```

```python
img = np.zeros((400,400,3), dtype=np.uint8)

pts = np.array([
    [200,50],
    [100,300],
    [300,300]
], np.int32)

cv2.polylines(
    img,
    [pts],
    True,
    (0,255,255),
    3
)

show(img)
```

```python
img = np.zeros((400,400,3), dtype=np.uint8)

pts = np.array([
    [200,50],
    [100,120],
    [130,280],
    [270,280],
    [300,120]
], np.int32)

cv2.polylines(img,[pts],True,(255,255,255),3)

show(img)
```

```python
img = np.zeros((400,600,3), dtype=np.uint8)

cv2.putText(
    img,
    "OpenCV",
    (100,200),
    cv2.FONT_HERSHEY_SIMPLEX,
    2,
    (255,255,255),
    3
)

show(img)
```

```python
img = np.zeros((400,400,3), dtype=np.uint8)

cv2.arrowedLine(
    img,
    (50,200),
    (350,200),
    (0,255,0),
    4
)

show(img)
```

```python
img = np.zeros((500,500,3), dtype=np.uint8)

cv2.line(img,(50,50),(450,50),(0,255,0),3)

cv2.rectangle(img,(50,100),(200,250),(255,0,0),3)

cv2.circle(img,(350,180),70,(0,0,255),3)

cv2.ellipse(img,(250,380),(100,50),45,0,360,(255,255,0),3)

cv2.putText(img,"OpenCV",(150,480),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,(255,255,255),2)

show(img)
```
