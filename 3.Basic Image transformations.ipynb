```python
import cv2
from matplotlib import pyplot as plt

# Read image
img = cv2.imread("wallp.png")

# OpenCV reads in BGR format
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display image
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
import cv2

img = cv2.imread("wallp.png")

# Height, Width, Channels
print("Shape:", img.shape)

# Total pixels
print("Total Pixels:", img.size)

# Data type
print("Data Type:", img.dtype)
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("wallp.png")

# Resize image
resized = cv2.resize(img, (300, 300))

plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.title("Resized Image")
plt.axis("off")
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("wallp.png")

# Rotate 90 degrees clockwise
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("wallp.png")

# Crop image
crop = img[400:600, 400:600]

plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("wallp.png")

# Draw rectangle
cv2.rectangle(img, (100,100), (300,300), (0,234,0), -1)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Rectangle")
plt.axis("off")
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("wallp.png")

# Draw circle
cv2.circle(img, (800,600), 250, (0,0,0), -1)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Circle")
plt.axis("off")
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("wallp.png")

# Add text
cv2.putText(img,
            "OpenCV",
            (800,600),
            cv2.FONT_HERSHEY_SIMPLEX,
            5,
            (0,255,0),
            3)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Text Added")
plt.axis()
plt.show()
```
